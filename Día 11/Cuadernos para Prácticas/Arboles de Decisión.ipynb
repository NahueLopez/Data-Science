```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree 
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df.species.unique() #Nos trae 1 de cada dato
```

```python
X = df.drop("species", axis=1) #Sacamos la columna species
X
```

```python
le = LabelEncoder()
especies = le.fit_transform(df["species"]) #Transforma las especies en numeros para poder utilizarlo
especies
```

```python
y = especies
y
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X,
                                                            y,
                                                            train_size=0.8,
                                                            random_state=42
                                                           )
```

```python
arbol = DecisionTreeClassifier()
```

```python
arbol.fit(X_entrena, y_entrena)
```

```python
plt.figure(figsize=(20,20))

plot_tree(decision_tree=arbol);
```

```python
plt.figure(figsize=(20,20))
plot_tree(decision_tree=arbol,
         filled=True #Colores
         );
```

```python
plt.figure(figsize=(20,20)) #Separacion y tamaÃ±o
plot_tree(decision_tree=arbol,
          class_names=["setosa","versicolor","virginica"], #Agrega la clase en el cuadro
         filled=True #Colores
         );
```

```python
plt.figure(figsize=(20,20)) #Separacion y tamaÃ±o
plot_tree(decision_tree=arbol,
          class_names=["setosa","versicolor","virginica"], #Agrega la clase en el cuadro
          feature_names=df.columns.to_list(),
          filled=True #Colores
         );
```
