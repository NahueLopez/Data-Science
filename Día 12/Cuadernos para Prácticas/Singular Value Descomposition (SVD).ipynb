```python
iris = load_iris()
X = iris.data
```

```python
X_centrado = X - np.mean(X, axis=0)
```

```python
U, sigma, VT = np.linalg.svd(X_centrado)
```

```python
k = 2
X_transformado = U[:, :k] * sigma[:k] #Se descompone los datos para disminuir la singularidad(Que sean mas parejos)
```

```python
especies = ["setosa","versicolor","virginica"]
```

```python
plt.figure(figsize=(8,6))
for i in range(3):
    plt.scatter(X_transformado[iris.target == i, 0],
                X_transformado[iris.target == i, 1],
                label=especies[i]
                )
plt.xlabel("Componente Principial 1")
plt.ylabel("Componente Principial 2")
plt.legend()
plt.title("Dataset iris transformado por SVD");
```
