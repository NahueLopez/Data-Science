```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris() #Creamos un diccionario con los datos de iris
iris
```

```python
X = iris.data #Todos los datos
y = iris.target #Los nombres
```

```python
X
```

```python
y
```

```python
X_centrado = X - np.mean(X, axis=0) #Le resta el promedio
```

```python
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X_centrado)
```

```python
especies = ["setosa","versicolor","virginica"]
```

```python
plt.figure(figsize=(8,6))
for i in range(0,3): #hace un fordes desde el 0 al 2
    plt.scatter(X_pca[y == i, 0], X_pca[y == i,1], label=especies[i])
plt.xlabel("Primer componente Principal")
plt.ylabel("Segundo componente Principal")
plt.legend()
plt.title("PCA del conjunto Iris");
```
