```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
np.random.seed(42) #Establece el numero en el que comienza a ser random
data = np.random.rand(100,2) # Genera una matriz de numero aletorios parejos de 0 a 1
data
```

```python
kmeans = KMeans(n_clusters=5, random_state=42) #Cantidad de cluster a hacer
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
centroides
```

```python
etiquetas = kmeans.labels_ #Una array en la cual determina a los 100 elementos para que cluster va
etiquetas
```

```python
plt.figure(figsize=(12,6))
plt.scatter(data[:, 0],
           data[:, 1],
           c=etiquetas,
           cmap="viridis",
           marker="o")
plt.scatter(centroides[:, 0],
            centroides[:,1],
            c="red",
            marker="x",
            s=220,
            label="Centroides"
           )
plt.title("Visualizacion de K-Means Clustering")
plt.legend();
```

```python
pinguinos = sns.load_dataset("penguins")
```

```python
pinguinos.dropna(inplace=True)
```

```python
data = pinguinos[["bill_length_mm", "bill_depth_mm"]]
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
etiquetas = kmeans.labels_
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data=pinguinos,
                x="bill_length_mm",
                y="bill_depth_mm",
                hue=etiquetas,
                palette="viridis"
               )
plt.scatter(
            centroides[:,0], 
            centroides[:,1],
            c="red",
            s=100,
            marker="x",
            label="Centroides"
            )
plt.title("K-Means Clustering en el conjunto de datos de Pinguinos")
plt.legend()
plt.xlabel("Longuitdo del Pico (mm)")
plt.ylabel("Profundidad del Pico (mm)");

```
