```python
x = [1,2,3,5,5,6,7]
y = [1,6,3,4,5,6,7]
```

```python
plt.plot(x,y)
plt.grid()
```

```python
plt.plot(x,y, "ro-") #el ro- hace roja la linea y la union de los puntos pone un punto
plt.grid()
```

```python
plt.plot(x,y, "b*:") #el b*: hace blue la linea y la union de los puntos pone un * y nteada
plt.grid()
```

```python
plt.plot(x,y, "b*:") #el b*: hace blue la linea y la union de los puntos pone un * y nteada
plt.axis([0,8,0,8])
plt.grid()
```
