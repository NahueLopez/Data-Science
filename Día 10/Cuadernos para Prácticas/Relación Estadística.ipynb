```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip"
           );
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip",
            kind="line"
           );
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip",
        hue="smoker",
           );
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip",
            hue="smoker",
            style="smoker",
           );
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip",
            hue="smoker",
            style="smoker",
            size="size",
           );
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip",
            hue="smoker",
            style="smoker",
            size="size",
            col="time"
           );
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip",
            hue="smoker",
            style="smoker",
            size="size",
            col="time",
            legend=None
           );
```
