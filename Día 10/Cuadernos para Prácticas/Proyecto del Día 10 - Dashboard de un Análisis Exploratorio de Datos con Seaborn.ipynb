```python
## Relaciones entre variables numéricas (por ejemplo, precio vs. peso en quilates)
sns.set(style="whitegrid")

graf = sns.relplot(data=diamantes, #Datos a graficar
                  x="price",       #Precio en X
                  y="carat",       #Corte en y
                  kind="scatter",  #Formato de grafico
                  height=6,        #Alto
                  aspect=1,      #Relación de aspecto (ancho/alto)
                  hue="cut",       #Cambiar colores segun el corte
                  style="cut"      #Cambiar marcadores segun el corte
                  )
#Cambio de labels
graf.fig.suptitle("Relación entre el Precio en Quilates y el Precio de los Diamantes", #Titulo
                  va="baseline", #Formato
                  ha="center"    #Formato
                 )
graf.set_axis_labels("Precio", "Peso en Quilates"); #Cambio de nombres eje x e y
```

```python
## Distribuciones de variables numéricas (por ejemplo, precio, peso en quilates)

graf2 = sns.displot(data=diamantes, #Datos a analizar
                  x="price",        #Eje x en precio
                  kind="hist",      #Formato de grafico
                  kde=True,         #Agrega una linea de media
                  color="violet",   #Color violeta
                  height=4,         #Alto del grafico
                  aspect=2,         #Relacion de alto y ancho
                )

graf2.fig.suptitle("Distribución del Precio de los Diamantes",  #Titulo
                   va="baseline",
                   ha="center"
                  )

graf2.set_axis_labels("Precio($)","Frecuencia");
```

```python
## Distribuciones de variables numéricas (por ejemplo, precio, peso en quilates)
g = sns.histplot(data=diamantes,
                x="carat",
                kde=True,
                color="violet"
                )
g.set_title("Distribución del Peso en Quilatess")
g.set(xlabel="Peso en Quilates", ylabel="Frecuencia");
```

```python
## Comparaciones de variables numéricas agrupadas por categorías (por ejemplo, calidad del corte)

g = sns.catplot(data=diamantes, #Datos
                  x="cut",      #Corte eje X
                  y="price",    #Precio eje Y
                  kind="box",   #Formato de grafico
                  height=5,     #Tamaño
                  aspect=2,     #Aspecto Alto,Ancho
                  order=["Fair","Good","Very Good","Premium","Ideal"] #Orden que se va a mostrar
                )
# Titulo
g.fig.suptitle("Comparacion del Precio de los Diamantes por Calidad de corte",
               va="baseline",
               ha="center"
                )
#Etiquetas EjesS
g.set_axis_labels("Calidad del Corte",
                  "Precio($)"
                 );
```

```python
#4.Utilizar jointplot() para explorar relaciones específicas entre pares de variables, incluyendo información estadística adicional
graf = sns.jointplot(data=diamantes,    # Datos a graficar
                  x="depth",            # Eje X: Profundidad del diamante
                  y="price",            # Eje Y: Precio del diamante
                  kind="kde",           # Tipo de gráfico: KDE (Kernel Density Estimate)
                  fill=True,            # Rellenar las áreas bajo las curvas KDE
                  space=0,              # Espacio entre los gráficos marginales y el gráfico principal
                  color="violet"        # Color del gráfico
                )
#Etiquetas de los ejes
graf.set_axis_labels("Profundidad del Diamante (%)",
                     "Precio ($)"
                    )

# Establecer el título del gráfico
graf.fig.suptitle("Relación entre la Profundidad y el Precio de los Diamantes",
                  va="baseline",   # Alineación vertical del título
                  ha="center",     # Alineación horizontal del título
                  fontsize=16      # Tamaño de fuente del título
                 );
```

```python
#5.Emplear pairplot() para visualizar las relaciones entre todas las variables numéricas del dataset de manera conjunta
g = sns.pairplot(diamantes,  # Datos a graficar
                corner=True,  # Mostrar solo la mitad inferior del gráfico para evitar duplicados
                )


# Titutlo
g.fig.suptitle("Relaciones entre Variables Numéricas en el Dataset de Diamantes",
              va="baseline",
               ha="center",
               fontsize=16
              );
```
