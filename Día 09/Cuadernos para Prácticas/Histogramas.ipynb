```python
numeros = np.random.randn(1000)
numeros
```

```python
plt.hist(numeros);
```

```python
plt.hist(numeros, bins=40); #bins cantidad de barras!
```

```python
plt.hist(numeros, bins=40,alpha=0.5); #alpha que tan transparente el grafico
```

```python
plt.hist(numeros, bins=40,alpha=0.5, color="red"); #color, color del grafico
```

```python
plt.hist(numeros, bins=40,alpha=0.5, color="red", edgecolor="green"); #edgecolor,color del borde de cada barra 
```

```python
plt.hist(numeros, bins=40,alpha=0.5, color="red", edgecolor="green", histtype="step"); #histtype, solamente muestra el contorno
```

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(500)
x3 = np.random.randn(100)

plt.hist(x1,histtype="stepfilled",alpha=0.3, bins=40);
plt.hist(x2,histtype="stepfilled",alpha=0.3, bins=40);
plt.hist(x3,histtype="stepfilled",alpha=0.3, bins=40);
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "frutas": ['Manzanas', 'Bananas', 'Cerezas', 'Duraznos'],
    "cantidades": [40, 25, 15, 20]
})

x = df["frutas"]
y = df["cantidades"]

plt.bar(x,y)
plt.title('Distribución de frutas en el almacén')
```
