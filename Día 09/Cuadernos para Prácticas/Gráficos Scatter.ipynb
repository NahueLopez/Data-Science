```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 10, 30) #Busca 30 numeros que tengna la misma distancia y eso numeros estaran entre 0 y 10
y = np.sin(x)
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y, "o"); #el o representa el tipo de punto que tendra
```

```python
aleatorio = np.random.RandomState(0)
marcadores = ["o","x","+","v","s","d"] #Diferentes tipos de marcados (como arriba es el o, estos son diferentes)
```

```python
for marcador in marcadores:
    plt.plot(aleatorio.rand(5), aleatorio.rand(5),marcador) #mostramos diferentes marcadores.
```

```python
plt.plot(x,y, ":ob");
```

```python
plt.plot(x,y, "-p");
```

```python
plt.plot(x,y, "-p", color="red");
```

```python
plt.plot(x,y, "-p", color="red", markersize=15);
```

```python
plt.plot(x,y, "-p", color="red", markersize=15, linewidth=4);
```

```python
plt.plot(x,y, "-p", color="red", markersize=15, linewidth=4, markerfacecolor="white");
```
